# 3. Limpando os dados: brasil_micro

```python
# Eliminando os combustíveis desnecessários para a análise:
brasil_micro.drop(["GLP", "GNV", "OLEO DIESEL", "OLEO DIESEL S10"], axis=0, inplace=True)
```

```python
#Eliminando as colunas desnecessárias para a análise:
brasil_micro.drop(["UNIDADE DE MEDIDA", "MARGEM MÉDIA REVENDA", "PREÇO MÉDIO DISTRIBUIÇÃO", "DESVIO PADRÃO DISTRIBUIÇÃO", "PREÇO MÍNIMO DISTRIBUIÇÃO", "PREÇO MÁXIMO DISTRIBUIÇÃO", "COEF DE VARIAÇÃO DISTRIBUIÇÃO"], axis=1, inplace=True)
```

```python
#Eliminando os meses/anos desnecessários para a análise:

#Filtrando o período de julho:
brasil_micro_julho = brasil_micro[brasil_micro['MÊS'] == 'jul.-22']
```

```python
#Filtrando o período de agosto:
brasil_micro_ago = brasil_micro[brasil_micro['MÊS'] == 'ago.-22']
```

```python
#Fazendo a junção dos dois períodos (julho + agosto) em um único dataframe:
brasil_micro = pd.merge(brasil_micro_julho, brasil_micro_ago, how='outer')
```
